```python
smd.columns
```

```python
#pre-processing smd

smdsmd['cast'] = smd['cast'].apply(literal_eval)
smd['crew'] = smd['crew'].apply(literal_eval)
smd['keywords'] = smd['keywords'].apply(literal_eval)
```

```python
smd['len_of_cast']=smd['cast'].apply(lambda x:len(x))

#selecting only director from crew
def get_director(x):
    for i in x:
        if i['job'] == 'Director':
            return i['name']
    return np.nan

smd['director'] = smd['crew'].apply(get_director)
```

```python
#considering only first 3 members from the cast

smd['cast'] = smd['cast'].apply(lambda x: [i['name'] for i in x] if isinstance(x, list) else [])
smd['cast'] = smd['cast'].apply(lambda x: x[:3] if len(x) >=3 else x)
smd['keywords'] = smd['keywords'].apply(lambda x: [i['name'] for i in x] if isinstance(x, list) else [])
```

```python
#removing spaces between names to not confuse and consider them as single ones 
smd['cast'] = smd['cast'].apply(lambda x: [str.lower(i.replace(" ", "")) for i in x])
smd['director'] = smd['director'].astype('str').apply(lambda x: str.lower(x.replace(" ", "")))

#giving more preference to director 
smd['director'] = smd['director'].apply(lambda x: [x,x, x])
```

```python
smd['director'][0]
```

```python
s = smd.apply(lambda x: pd.Series(x['keywords']),axis=1).stack().reset_index(level=1, drop=True)
s.name = 'keyword'
s = s.value_counts()
s[:5]
```

```python
#removing keywords which occur less than once 
s = s[s > 1]
def modify_keywords(x):
    words = []
    for i in x:
        if i in s:
            words.append(i)
    return words
smd['keywords'] = smd['keywords'].apply(modify_keywords)
```

```python
#stemming keywords using Stemmer

stemmer = SnowballStemmer('english')
smd['keywords'] = smd['keywords'].apply(lambda x: [stemmer.stem(i) for i in x])
smd['keywords'] = smd['keywords'].apply(lambda x: [str.lower(i.replace(" ", "")) for i in x])
```

```python
#combining the features 
smd['final']=smd['keywords'] + smd['cast'] + smd['director'] + smd['genres']
smd['final'] = smd['final'].apply(lambda x: ' '.join(x))
smd['final'][0]
```

```python
#cosine similarity
cv = CountVectorizer(analyzer='word',ngram_range=(1, 2),min_df=0, stop_words='english')
count_matrix = cv.fit_transform(smd['final'])
sim_scores = cosine_similarity(count_matrix, count_matrix)
```

```python
smd=smd.reset_index()
titles = smd['title']
titles.head(3)
```

```python
reader = Reader()
data = Dataset.load_from_df(ratings[['userId', 'movieId', 'rating']], reader)
```

```python
#using svd algorithm for matrix factorization and running 5-fold cross-validation

algo = SVD()
cross_validate(algo, data, measures=['RMSE', 'MAE'], cv=5, verbose=True)
```

```python
trainset = data.build_full_trainset()
algo.fit(trainset)
```

```python
def convert_int(x):
    try:
        return int(x)
    except:
        return np.nan
```

```python
id_data = pd.read_csv('links_small.csv')[['movieId', 'tmdbId']]
id_data['tmdbId'] = id_data['tmdbId'].apply(convert_int)
id_data.columns = ['movieId', 'id']
id_data = id_data.merge(smd[['title', 'id']], on='id').set_index('title')
indices_data = id_data.set_index('id')
```

```python
def hybrid_recommender(userId, title):
    idx = indices[title]
    tmdbId = id_data.loc[title]['id']
    movie_id = id_data.loc[title]['movieId']
    similarity_scores = list(enumerate(sim_scores[int(idx)]))
    similarity_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)
    similarity_scores = similarity_scores[1:31]
    movie_indices = [i[0] for i in similarity_scores]
    movies = smd.iloc[movie_indices][['title', 'vote_count', 'vote_average', 'release_date', 'id']]
    movies['est_rating'] = movies['id'].apply(lambda x: algo.predict(userId, indices_data.loc[x]['movieId']).est)
    movies = movies.sort_values('est_rating', ascending=False)
    return movies.head(10)
```
